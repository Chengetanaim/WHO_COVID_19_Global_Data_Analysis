## Data Cleaning

Before thinking of how I can deal with these null values I create a new dataframe which contains these null values.

```python
contains_nulls = df[df.Country_code.isna()]
```

```python
contains_nulls
```

Upon observing the null values dataframe, I see that Namibia is the only country whose Country_code is being recorded as null.<br>
What could be the reason for this?  How can I rectify this?<br>
What comes to my mind immediately is that maybe the person recording this data did not know the Country_code for Namibia.<br>
These null values are considered to be missing completely at random.<br><br>
"Missing data are MCAR when the probability of missing data on a variable is unrelated to any other measured variable and is unrelated to the variable with missing values itself. In other words the missingness on the variable is completely unsystematic. For example when data are missing for respondents for which their questionnaire was lost in the mail. This assumption can be tested by separating the missing and the complete cases and examine the group characteristics. If characteristics are not equal for both groups, the MCAR assumption does not hold."<br> 
https://www.missingdata.nl/missing-data/missingdatamechanisms/mcar/#:~:text=Missing%20completely%20at%20random%20(MCAR,variable%20with%20missing%20values%20itself.

```python
df.loc[:, 'Country_code'].fillna('NA', inplace=True)
```

I then replace the missing values with Namibia's country code which is NA.<br>
Note that pandas may have recorded these values as NAN since Namibia's country code is NA which pandas may have took as Not Available.  (this has got to be the most logical reason.)

```python
df.info()
```

Whatever the reason was for the null values, our dataframe no longer has them.<br>
Now on to binning...

```python
df = df.assign(
    New_cases_binned = pd.cut(df.New_cases, bins=3, labels=['low', 'med', 'high']),
    New_deaths_binned = pd.cut(df.New_deaths, bins=3, labels=['low', 'med', 'high']),
)
```

The code in the above cell is performing a data enrichment process where I add new columns New_cases_binned and New_deaths_binned to the dataframe.<br>
This puts the New_cases and New_deaths columns into bins named low, med and high such that we can easily query the days when new covid cases or deaths  high.

```python
df[df.New_cases_binned=='high'].sort_values(by='New_cases', ascending=False)
```

In December 2022 there are days when new covid cases found were excessively high.

```python
df[df.New_cases_binned=='low'].sort_values(by='New_cases', ascending=False)
```

Upon viewing the result I observe that there are some instances where New_cases was recorded as being below 0.<br>
How is this possible?  How can new covid cases per day be recorded as negative?

```python
df[(df.New_cases<0) | (df.Cumulative_cases<0) | (df.New_deaths<0) | (df.Cumulative_deaths<0)].sort_values(by='New_cases').head()
```

```python
df[(df.New_cases<0) | (df.Cumulative_cases<0) | (df.New_deaths<0) | (df.Cumulative_deaths<0)].sort_values(by='New_cases').shape[0]
```

We have 93 rows which contains at least a single negative value.

```python
df.loc['2023-02'].query('Country=="China"').head()
```

I query one random month for China where I saw these negative values being recorded.<br>
The new covid cases found on the first of February are said to be -1073379, but on the following day there are 1138221.<br>
The trend for the opening 5 days of February in 2023 is somewhat the same.<br>
On the other hand the cumulative cases have also been affected.  On 3 February 2023 the cumulative cases dropped because the new cases recorded the day before were negative.<br>
Not really sure what could be the reason for these negative values, but I then conclude that these values are a human error, especially given that we only have a small portion where values are recorded as negative.<br>
The next logical step is to change the negative values into positive, and update the columns Cumulative_cases, Cumulative_deaths, New_cases_binned and New_deaths_binned as they all depend on the columns with the negative values.  (New_cases and New_deaths)

```python
df = df.assign(
    New_cases=lambda x: abs(x.New_cases),
    New_deaths=lambda x: abs(x.New_deaths),
)
df['Cumulative_cases'] = df[['Country','New_cases']].groupby('Country').cumsum()
df['Cumulative_deaths'] = df[['Country','New_deaths']].groupby('Country').cumsum()
df = df.assign(
    New_cases_binned = pd.cut(df.New_cases, bins=3, labels=['low', 'med', 'high']),
    New_deaths_binned = pd.cut(df.New_deaths, bins=3, labels=['low', 'med', 'high']),
    Cumulative_cases_binned = lambda x: pd.cut(x.Cumulative_cases, bins=3, labels=['low', 'med', 'high']),
    Cumulative_deaths_binned = lambda x: pd.cut(x.Cumulative_deaths, bins=3, labels=['low', 'med', 'high'])
)
```

The issue has now been rectified.
